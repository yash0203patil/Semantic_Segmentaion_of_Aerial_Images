```python
from tensorflow import keras
import random
import matplotlib.pyplot as plt
import pickle
import numpy as np

model = keras.models.load_model("semantic_segmentation_aerial_satelite_images.h5")
x_test = pickle.load(open('X_test.pickle', 'rb'))
y_test = pickle.load(open('y_test.pickle', 'rb'))
```

```python
sample_images = random.randint(0, 265)
image = x_test[sample_images]
ground_truth = np.argmax(y_test, axis=3)[sample_images]

image_input = np.expand_dims(image, 0)
prediction = model.predict(image_input)
predicted_image = np.argmax(prediction, axis=3)[0,:,:]
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(231)
plt.title('Testing Image')
plt.imshow(image)
plt.subplot(232)
plt.title('Testing Label')
plt.imshow(ground_truth)
plt.subplot(233)
plt.title('Prediction on test image')
plt.imshow(predicted_image)
plt.show()
```
